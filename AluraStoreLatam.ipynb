#1. Análisis de facturación



Ingreso total por cada tienda

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
print("Resumen de Ingresos por Tienda")

# 1. Almacenamos los ingresos en un diccionario para poder compararlos
ingresos_por_tienda = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}

# 2. Mostramos el resumen de cada tienda usando el diccionario
for nombre_tienda, total_ingreso in ingresos_por_tienda.items():
    print(f'Ingreso Total {nombre_tienda}: $ {total_ingreso:,.0f}')

# 3. Encontramos la tienda con el ingreso MÁXIMO
#    Usamos max() con un 'key' para que se fije en los valores del diccionario
tienda_mayor_venta = max(ingresos_por_tienda, key=ingresos_por_tienda.get)
mayor_ingreso = ingresos_por_tienda[tienda_mayor_venta]

# 4. Encontramos la tienda con el ingreso MÍNIMO
#    Funciona igual que max()
tienda_menor_venta = min(ingresos_por_tienda, key=ingresos_por_tienda.get)
menor_ingreso = ingresos_por_tienda[tienda_menor_venta]

# 5. Imprimimos el análisis de facturación
print(f"Tienda con mayor facturación: {tienda_mayor_venta} (${mayor_ingreso:,.0f})")
print(f"Tienda con menor facturación: {tienda_menor_venta} (${menor_ingreso:,.0f})")
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Creamos un diccionario con los nombres y los DataFrames de las tiendas.
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

# Recorremos cada tienda para analizarla
for nombre_tienda, df_tienda in tiendas.items():
    print(f"--- Ventas por Categoría: {nombre_tienda} ---")

    ventas_por_categoria = df_tienda.groupby('Categoría del Producto')['Precio'].count().sort_values(ascending=False)
    print("Cantidad de ventas por categoría:")
    print(ventas_por_categoria.to_frame(name='Cantidad de Ventas'))
    # --- MOSTRANDO LA CATEGORÍA MÁS POPULAR ---
    categoria_mas_popular = ventas_por_categoria.idxmax()
    cantidad_ventas = ventas_por_categoria.max()
    print(f"\nLa categoría más popular es: '{categoria_mas_popular}' con {cantidad_ventas} ventas.")

    # Agregamos un espacio para separar los resultados de cada tienda
    print("\n" + "="*50 + "\n")
```

# 3. Calificación promedio de la tienda


Valoración media por tienda

En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python

# Diccionario para organizar los DataFrames
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

# Diccionario vacío para guardar el promedio general de cada tienda y compararlos después
calificaciones_generales = {}

# Recorremos cada tienda para analizarla
for nombre_tienda, df_tienda in tiendas.items():
    print(f"--- Análisis de Calificaciones: {nombre_tienda} ---")
    # 1. Calcular el promedio GENERAL para la tienda actual
    promedio_general = df_tienda['Calificación'].mean()
    calificaciones_generales[nombre_tienda] = promedio_general  # Guardamos el resultado
    print(f"Calificación promedio general: {promedio_general:.2f}\n")

    # 2. Calcular el promedio POR CATEGORÍA
    #    Agrupamos por categoría, calculamos la media y ordenamos de mayor a menor.
    promedio_por_categoria = df_tienda.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)

    print("Calificación promedio por categoría:")
    print(promedio_por_categoria.to_frame(name='Calificación Promedio').round(2))

    # Separador para la siguiente tienda
    print("\n" + "="*50 + "\n")


# --- ANÁLISIS FINAL: ¿QUÉ TIENDA TIENE MEJOR Y PEOR CALIFICACIÓN? ---
# Ahora usamos el diccionario que fuimos llenando para encontrar el máximo y mínimo

tienda_mejor_calificada = max(calificaciones_generales, key=calificaciones_generales.get)
tienda_peor_calificada = min(calificaciones_generales, key=calificaciones_generales.get)

print("--- Resumen General de Satisfacción del Cliente ---")
print(f"Tienda con MEJOR calificación: {tienda_mejor_calificada} ({calificaciones_generales[tienda_mejor_calificada]:.2f} estrellas)")
print(f"Tienda con PEOR calificación: {tienda_peor_calificada} ({calificaciones_generales[tienda_peor_calificada]:.2f} estrellas)")
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
# Diccionario para organizar los DataFrames
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

# Recorremos cada tienda para analizarla
for nombre_tienda, df_tienda in tiendas.items():
    print(f"--- Análisis de Ventas por Producto: {nombre_tienda} ---")
    # 1. Contar las ventas de cada producto.
    conteo_productos = df_tienda['Producto'].value_counts()
    # 2. Obtener el producto MÁS vendido (el primero de la lista)
    producto_mas_vendido_nombre = conteo_productos.index[0]
    producto_mas_vendido_cantidad = conteo_productos.iloc[0]
    # 3. Obtener el producto MENOS vendido (el último de la lista)
    producto_menos_vendido_nombre = conteo_productos.index[-1]
    producto_menos_vendido_cantidad = conteo_productos.iloc[-1]
    # 4. Encontrar la categoría de cada uno de esos productos
    categoria_mas_vendido = df_tienda[df_tienda['Producto'] == producto_mas_vendido_nombre]['Categoría del Producto'].iloc[0]
    categoria_menos_vendido = df_tienda[df_tienda['Producto'] == producto_menos_vendido_nombre]['Categoría del Producto'].iloc[0]
    # 5. Imprimir los resultados
    print(f"Producto MÁS vendido: '{producto_mas_vendido_nombre}'")
    print(f"   - Categoría: {categoria_mas_vendido}")
    print(f"   - Cantidad vendida: {producto_mas_vendido_cantidad}\n")

    print(f"Producto MENOS vendido: '{producto_menos_vendido_nombre}'")
    print(f"   - Categoría: {categoria_menos_vendido}")
    print(f"   - Cantidad vendida: {producto_menos_vendido_cantidad}")

    # Separador para la siguiente tienda
    print("\n" + "="*50 + "\n")
```

# 5. Envío promedio por tienda

Valor del envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
# Diccionario para organizar los DataFrames
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

# Diccionario para guardar los resultados y poder compararlos al final
costos_promedio = {}

print("--- Costo de Envío Promedio por Tienda ---")

# Recorremos cada tienda para analizarla
for nombre_tienda, df_tienda in tiendas.items():
    # 1. Calcular el costo de envío promedio para la tienda actual
    costo_promedio = df_tienda['Costo de envío'].mean()
    # 2. Guardar el resultado para el análisis final
    costos_promedio[nombre_tienda] = costo_promedio
    # 3. Imprimir el resultado formateado (con separador de miles y 2 decimales)
    print(f'Para {nombre_tienda}, el costo de envío promedio es: $ {costo_promedio:,.2f}')

# --- ANÁLISIS FINAL: ¿QUÉ TIENDA TIENE EL ENVÍO MÁS CARO Y MÁS BARATO? ---
tienda_envio_mas_caro = max(costos_promedio, key=costos_promedio.get)
tienda_envio_mas_barato = min(costos_promedio, key=costos_promedio.get)
print("\n--- Resumen de Costos de Envío ---")
print(f"Tienda con el envío más CARO: {tienda_envio_mas_caro} (${costos_promedio[tienda_envio_mas_caro]:,.2f} en promedio)")
print(f"Tienda con el envío más BARATO: {tienda_envio_mas_barato} (${costos_promedio[tienda_envio_mas_barato]:,.2f} en promedio)")
```
